```python
import matplotlib.pyplot as plt
import pandas as pd 

#tabela = pd.read_excel("Vendas.xlsx")
grafico = tabela[['Produto','Valor Final']].head(10)
display(grafico)

# Plotar o gráfico de barras
plt.figure(figsize=(10, 6))
plt.barh(grafico['Produto'], grafico['Valor Final'], color='skyblue')
plt.xlabel('Valor Final')
plt.ylabel('Produto')
plt.title('Valor Final por Produto')
plt.gca().invert_yaxis()  # Inverter os produtos no eixo y para que o produto mais caro esteja no topo
plt.tight_layout()
plt.show()
```
